```
import numpy as np
import matplotlib.pyplot as plt

# dumy data kamar tidur
bedrooms = np.array([1, 1, 2, 2, 2, 3, 4, 4, 5, 5])

# dumy harga rumah
harga_rumah = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])

# menampilkan scater plot
%matplotlib inline
plt.scatter(bedrooms, harga_rumah)

from sklearn.linear_model import LinearRegression
bedrooms = bedrooms.reshape(-1, 1)
linreg = LinearRegression()
linreg.fit(bedrooms, harga_rumah)

plt.plot(bedrooms, linreg.predict(bedrooms))
```
